### Believe Results
What key metrics help you trust results:

- Sample size
- Who/what was sampled? Are they relavant to the question?
- How was the survey conducted

Example
Here is a brief description of how the test worked, to help you follow the rest of the lesson. The test consisted of three parts:

Users were shown 12 photos in the first part.
Users were shown another 12 photos in the second part.
Users were shown 48 photos in the third part and asked if they saw each photo in the first part, the second part, or neither.
After the test, users were given two scores:

A "Recognition score", calculated as the percentage of times they correctly identified whether they saw the face at all, regardless of which part the face was from.
A "Temporal memory score", calculated as the percentage of recognized faces that were identified with the correct part (part 1 or part 2).
Users were advised to take a 5 minute break between each part of the test.

Recognition Score: Recognize faces you saw before
Temporal Memory Score: Remember when you saw each face

### Causal Inference

Correlation does not imply causation.

Golden Arches Theory (of conflict prevention):
No two countries with a McDonald's have ever gone to war since opening the McDonald's. (Please note that we are not endorsing Friedman's theory. Rather, we are trying to demonstrate that just because two variables are related, this doesn't mean that one causes the other to occur.)

Show relationships (e.g. scatterplots) -> Observational Studies & Surveys

Show causation -> controlled experiment

### Central Tendency

Measures of center:  
Mode - value at which frequency is the highest (most common value)  
Median - value in the middle of the distribution  
Mean - average value of the distribution   

Uniform distribution - no mode, all values are relatively the same  
Multi-modal - distribution with more than one mode (two peaks)

mean sample:
$\bar{x} = (\sum_{i=0}^n x_i )/ n$

Mean population:
$\mu = (\sum x )/ N$

Median (n even):  
$(x_{n/2} + x_{((n/2) + 1)})/2$  

Median (n odd):  
$x_{(n+1)/2}$

Right skewed distribution:  
mode < median < mean  
Normal distribution:       
mode = median = mean

### Variability

Range is the difference between the maximum value and the minimum value observed

Q1 = 25%    
Q2 = 50%    
Q3 = 75%  

Interquartile Range (IQR)  
IQR = Q3 - Q1  (~50% of data)


Outlier Equation:  
$< Q_1 - 1.5*IQR$    
$> Q_3 + 1.5*IQR$  
        
Dealing with outliers:
Chop off the tails

Mean will not always be inside IQR, although it usually is. 

Deviation from mean:  
$x_i-\bar{x}$

Average Deviation from mean:    
$\Sigma_{i=0}^n|x_i-\bar{x}|/n$

Sum of squared deviation:   
$\Sigma_i(x_i - \bar{x})^2$

Variance (Averaged squared deviation):  

$\displaystyle \frac{\Sigma_i(x_i - \bar{x})^2}{n}$

Mean of squared deviations


Standard deviation (of population or dataset):

$\displaystyle \sigma$ = $\displaystyle \frac{\sqrt{\Sigma_i(x_i - \bar{x})^2}}{n}$

Square root of average squared deviation

Notes: 
- 68% of data falls within $1\sigma$ of the mean in a normal distribution
- 95% of data falls within $2\sigma$ of the mean in a normal distribution

Bessel's Correction (sample standard deviation):  
Using a sample to estimate the standard deviation of a population  
$\displaystyle s$ = $\displaystyle \frac{\sqrt{\Sigma_i(x_i - \bar{x})^2}}{n-1}$
